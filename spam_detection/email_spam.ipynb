```python

X=df["message_content"]

```

```python
print(X.dtype)
```

```python
y=df['is_spam']
```

```python
df["is_spam"].value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()

```

```python

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train_vect=vectorizer.fit_transform(X_train)
X_test_vect=vectorizer.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
```

```python

model=GaussianNB()
```

```python
from sklearn.metrics import accuracy_score,f1_score,confusion_matrix
```

```python

X_test_dense=X_test_vect.toarray()
X_train_dense=X_train_vect.toarray()
model.fit(X_train_dense, y_train)
y_pred=model.predict(X_test_dense)
print("accuray:",accuracy_score(y_test,y_pred))
print("f1 score:",f1_score(y_test,y_pred))
print("confusion matrices:",confusion_matrix(y_test,y_pred))
email=["Congratulations! You won a free iPhone. Click here to claim your prize.",
"Meeting at 3 PM. Please send me the report before then.",
"Get a free vacation now! Limited-time offer, claim fast.",
"Can we reschedule our call for tomorrow?",
"Urgent: Your bank account has been compromised. Click the link to secure it.",
"hurray!!you won a lottery ticket",
"we happy to annoce that you have me hirred by tcs",]
    
email_vect= vectorizer.transform(email)
email_dense=email_vect.toarray()
prediction=model.predict(email_dense)
print(prediction)
```
